```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import seaborn as sns
```

```python
df=pd.read_csv('Fremont_Bridge_Bicycle_Counter.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
df.columns=['Total','East','West']
```

```python
plt.style.use('seaborn')
```

```python
df.plot()
```

```python
df.info()
```

```python
df.resample('BQ').sum().plot()
```

```python
df.resample('w').sum().plot()
```

```python
ax=df.resample('D').sum().rolling(365).sum().plot()
#ax.set_ylim(0,None)
```

```python
df.groupby(df.index.time).mean().plot()
```

```python
pivot=df.pivot_table('Total',index=df.index.time,columns=df.index.date)
```

```python
pivot.iloc[:5,:5]
```

```python
pivot.plot(legend=False, alpha=0.01,figsize=(18, 18))
```

```python
pivot2=df.pivot_table('West',columns=df.index.date,index=df.index.time)
pivot2.head()
```

```python
pivot2.plot(legend=False,alpha=0.1,figsize=(12,12))
```
